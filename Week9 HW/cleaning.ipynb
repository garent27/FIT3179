```python
import pandas as pd
import json
from pathlib import Path

# Read CSV file
df = pd.read_csv("fastest_solve.csv")

# Read JSON file
with open("ne_110m.json") as f:
    topojson_data = json.load(f)

# Extract country names from TopoJSON
topo_countries = set()
for geometry in topojson_data["objects"]["ne_110m_admin_0_countries"]["geometries"]:
    topo_countries.add(geometry["properties"]["NAME"])

# Create country name mapping dictionary
country_mapping = {
    "Chinese Taipei": "Taiwan",
    "Hong Kong": "Hong Kong S.A.R.",
    "Macau": "Macao S.A.R",
    "Czech Republic": "Czechia",
    "Democratic Republic of the Congo": "Dem. Rep. Congo",
    "Dominican Republic": "Dominican Rep.",
    "Bosnia and Herzegovina": "Bosnia and Herz.",
    "United States": "United States of America"
}

# Apply the mapping to the DataFrame
df['Country'] = df['Country'].replace(country_mapping)

# Create a DataFrame with all countries from TopoJSON
all_countries_df = pd.DataFrame({'Country': list(topo_countries)})

# Merge with existing data, keeping all countries and filling missing values
merged_df = pd.merge(all_countries_df, df, on='Country', how='left')
merged_df['BestSeconds'] = merged_df['BestSeconds'].fillna('N/A')
merged_df['Person'] = merged_df['Person'].fillna('No Data')

# Save the complete dataset
merged_df.to_csv("fastest_solve_corrected.csv", index=False)

# Print summary
print(f"\nFinal Summary:")
print(f"Total countries in map: {len(topo_countries)}")
print(f"Countries with solve times: {len(df)}")
print(f"Total countries in final dataset: {len(merged_df)}")
```
